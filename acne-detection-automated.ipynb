```python
from ultralytics import YOLO
import cv2
import numpy as np
from PIL import Image
import requests
from io import BytesIO
import matplotlib.pyplot as plt

# 1. Load trained model
model = YOLO("Acne_Detection_2-1/yolo_model/acne_detection/weights/best.pt")

# 2. Download image from URL
image_url = "https://th.bing.com/th/id/OIP.H9Lg0C89BAF8ifquVX4aRgHaFL?cb=iwp2&rs=1&pid=ImgDetMain"
response = requests.get(image_url)
image = Image.open(BytesIO(response.content)).convert("RGB")
image_np = np.array(image)

# 3. Save temporarily to file (optional, if you want to use path-based inference)
image.save("test_image.jpg")

# 4. Run inference using path (recommended for YOLOv8)
results = model.predict(source="test_image.jpg", conf=0.4)

# 5. Visualize result
for r in results:
    im_bgr = r.plot()  # returns a BGR image with predictions
    im_rgb = cv2.cvtColor(im_bgr, cv2.COLOR_BGR2RGB)
    plt.imshow(im_rgb)
    plt.axis("off")
    plt.title("YOLOv8 Acne Detection")
    plt.show()
```
